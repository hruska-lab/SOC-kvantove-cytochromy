```python
!pip install rdkit
import rdkit
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.ipython_useSVG = True

def mol_with_atom_index( mol ):
    atoms = mol.GetNumAtoms()
    for idx in range( atoms ):
        mol.GetAtomWithIdx( idx ).SetProp( 'molAtomMapNumber', str( mol.GetAtomWithIdx( idx ).GetIdx() ) )
    return mol
```

```python
mol = Chem.MolFromSmiles( "C1=CC=C(C(=C1)CC(=O)O)NC2=C(C=CC=C2Cl)Cl" )
```

```python
#Default
mol
#With index
mol_with_atom_index(mol)
```

```python
!pip install py3Dmol
import py3Dmol
from rdkit import Chem
from rdkit.Chem import (
    AllChem,
    rdCoordGen,
)
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.ipython_useSVG=True  # Use higher quality images for molecules
import py3Dmol
```

```python
diclofenac = Chem.MolFromSmiles("C1=CC=C(C(=C1)CC(=O)O)NC2=C(C=CC=C2Cl)Cl")
rdCoordGen.AddCoords(diclofenac)
diclofenac
```

```python
def molecule_to_3d(molecule):
    mol = Chem.Mol(molecule)
    mol = AllChem.AddHs(mol, addCoords=True)
    AllChem.EmbedMolecule(mol)
    AllChem.MMFFOptimizeMolecule(mol)
    return mol

diclofenac_3d = molecule_to_3d(diclofenac)
```

```python
view = py3Dmol.view(
    data=Chem.MolToMolBlock(diclofenac_3d),  # Convert the RDKit molecule for py3Dmol
    style={"stick": {}, "sphere": {"scale": 0.3}}
)
view.zoomTo()
```

```python
view.render('png', '/path/to/folder/ex.png', dpi=1000, transparent=True)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
smi = "C1=CC=C(C(=C1)CC(=O)[O-])NC2=C(C=CC=C2Cl)Cl"
#smi = "C1=CC=C(C(=C1)CC(=O)O)NC2=C(C=CC=C2Cl)Cl"
mol = Chem.MolFromSmiles(smi)
mol = Chem.AddHs(mol)
print("atoms", mol.GetNumAtoms())
mol
```

```python
Draw.MolToFile(mol, 'diclofenac-2D.png')
```

```python
AllChem.EmbedMolecule(mol)
AllChem.MMFFOptimizeMolecule(mol, maxIters = 300)
mol
```

```python
Draw.MolToFile(mol, 'diclofenac-3D.png')
```
